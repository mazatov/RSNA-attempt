Create annotations for xray images in appropriate Yolov3 format. 

Yolov3 format: 
Row format: image_file_path box1 box2 ... boxN;

path/to/img1.jpg 50,100,150,200,0 30,50,200,120,3

path/to/img2.jpg 120,300,250,600,2
...

```python
import argparse
import io
import os
import numpy as np
```

```python
#data_path='data\players'
#data_path='data\players1'
data_path='players'
```

```python
data_path = os.path.expanduser(data_path)
```

```python
files = os.listdir(data_path)
```

```python
i=0
all_boxes=[]
for f in files:
    if f[-4:]=="jpeg" and os.path.isfile(os.path.join(data_path,f[:-5]+'.txt')) :
        with open(os.path.join(data_path,f[:-5]+'.txt')) as f1:
            lines = (line for line in f1)
            box = np.loadtxt(lines,  delimiter=" ",skiprows=1)
            all_boxes.append(box)
```

```python
i=0
output_file=open('train.txt','w')
for f in files:
    if f[-4:]=="jpeg" and os.path.isfile(os.path.join(data_path,f[:-5]+'.txt')) :        
        #Start writing a text files train.txt
        # component 1: data_path + image_name
        part1 = os.path.join(data_path,f)
        print(part1)
        # component 2: box data. 
        box=all_boxes[i]
        print(box.shape)
        if box.size>0:
            if len(box.shape) == 1:
                box = box.reshape((1,4))
            s=''
            for j in range(box.shape[0]):
                bj=box[j,:]
                sj=np.array2string(bj.astype(int),separator=',')            
                sj1=sj.replace(" ", "")
                #print(sj1)
                s = s + sj1[1:-1] + ',0 '
            str = part1 + ' ' + s[:-1]
            output_file.write(str)
            output_file.write('\n')
            #print(str)                        
        i=i+1
output_file.close()        
```
